## Step 1: 기관 정보 입력 (✏️)

```python
SCHOOL_INFO = {
    'name': '✏️ [교육 기업명/학원명]',
    'type': '✏️ [학원|학교|기업연수|온라인강의]',
    'subject': '✏️ [주요 과목/분야]',
    'students': '✏️ [수강생 수]',
    'teachers': '✏️ [강사/교사 수]',
}
print('✅ 설정 완료:', SCHOOL_INFO['name'])
```

## Step 2: AX 진단 (10개 문항)

```python
SCORES = {
    '교안/교재 자동화': 1,        # ✏️ 1~5
    '학습 레벨 자동 진단': 1,     # ✏️ 1~5
    '숙제/피드백 자동화': 1,      # ✏️ 1~5
    '성적 분석 자동화': 1,        # ✏️ 1~5
    '학습 추이 예측': 1,          # ✏️ 1~5
    'AI 튜터/챗봇': 1,            # ✏️ 1~5
    '학부모 커뮤니케이션 자동화': 1, # ✏️ 1~5
    '수업 품질 분석': 1,          # ✏️ 1~5
    '캘린더/스케줄 자동화': 1,    # ✏️ 1~5
    '통합 학습 대시보드': 1,      # ✏️ 1~5
}
total=sum(SCORES.values())
level='🟢 Starter' if total<25 else ('🔵 Growing' if total<40 else '🟣 AI-Ready')
print(f'총점: {total}/50점 — {level}')
```

## Step 3: 시각화

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots(figsize=(10,5))
ax.barh(list(SCORES.keys()),list(SCORES.values()),
        color=['#e74c3c' if v<=2 else '#f39c12' if v<=3 else '#2ecc71' for v in SCORES.values()])
ax.axvline(3,color='gray',linestyle='--',alpha=0.5)
ax.set_xlim(0,5)
ax.set_title(f'{SCHOOL_INFO["name"]} 교육 AX 진단')
plt.tight_layout()
plt.savefig('edu_ax_diagnosis.png',dpi=150,bbox_inches='tight')
plt.show()
```

## Step 4: 맞춤 교육 혁신 로드맵

```python
weak=[k for k,v in SCORES.items() if v<=2]
prompt=f"""교육 혁신 AI 컨설턴트로서 맞춤 전략을 작성하세요.
기관: {SCHOOL_INFO['name']} ({SCHOOL_INFO['type']}) — {SCHOOL_INFO['subject']}
진단 총점: {total}/50점 ({level})
즉시 개선 가능 항목: {', '.join(weak) if weak else '전문화/고도화 단계'}
12주 로드맵 포함:
1. 즉시 실행 3가지 (Gemini API 활용)
2. 학생/학부모 경험 개선 방안
3. 강사 업무 자동화 전략
형식: 마크다운, 교육 현장 맞춤형"""
resp=model.generate_content(prompt)
print(resp.text)
```

## Step 5: 다운로드

```python
with open('edu_ax_report.md','w',encoding='utf-8') as f:
    f.write(f'# {SCHOOL_INFO["name"]} 교육 AX 진단 리포트\n\n')
    f.write(f'**총점**: {total}/50 | **레벨**: {level}\n\n')
    f.write(resp.text)
from google.colab import files
files.download('edu_ax_report.md'); files.download('edu_ax_diagnosis.png')
print('✅ 완료! 다음: W2 교안 자동화')
```
