## Step 0: 환경 설정

```python
import subprocess
subprocess.run(['apt-get','-qq','-y','install','fonts-nanum'],capture_output=True)
import matplotlib.pyplot as plt, matplotlib.font_manager as fm
_n=[f for f in fm.findSystemFonts() if 'Nanum' in f]
if _n: fm.fontManager.addfont(_n[0]); plt.rcParams['font.family']=fm.FontProperties(fname=_n[0]).get_name()
plt.rcParams['axes.unicode_minus']=False
!pip install -q google-generativeai pandas matplotlib
try:
    from google.colab import userdata; API_KEY=userdata.get('GEMINI_API_KEY')
except: API_KEY=input('GEMINI_API_KEY: ')
import google.generativeai as genai; genai.configure(api_key=API_KEY)
model=genai.GenerativeModel('gemini-2.0-flash'); print('✅ 준비 완료')
```

## Step 1: 정보 입력 (✏️)

```python
INFO={'name':'✏️ [교육 기업명/학원명]','class':'✏️ [반/학년]'}
print('✅',INFO['name'])
```

## Step 2: 알림 유형 선택(성적/행사/공지) → Gemini 문자 초안 → 목록 CSV 출력

```python
import pandas as pd, numpy as np
# 샘플 데이터
import random; random.seed(42)
scores=[random.randint(60,100) for _ in range(20)]
scores[7]=15; scores[14]=18  # 이상치 보장
df=pd.DataFrame({'학생':[f'학생{i+1}' for i in range(20)],'점수':scores})
print(df.describe())
```

## Step 3: AI 분석

```python
p=f"""교육 데이터 분석가로서 알림 유형 선택(성적/행사/공지) → Gemini 문자 초안 → 목록 CSV 출력 컨텍스트로 분석하세요.
기관: {{INFO['name']}} {{INFO['class']}}
주요 인사이트 3가지 + 개선 방안. 마크다운."""
resp=model.generate_content(p); print(resp.text)
```

## Step 4: 다운로드

```python
with open('result.md','w',encoding='utf-8') as f: f.write(resp.text)
df.to_csv('data.csv',index=False,encoding='utf-8-sig')
from google.colab import files
files.download('result.md'); files.download('data.csv')
print('✅ 완료!')
```
