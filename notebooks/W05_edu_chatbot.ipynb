## Step 0: 환경 설정

```python
!pip install -q google-generativeai pandas matplotlib
try:
    from google.colab import userdata; API_KEY=userdata.get('GEMINI_API_KEY')
except: API_KEY=input('GEMINI_API_KEY: ')
import google.generativeai as genai; genai.configure(api_key=API_KEY)
model=genai.GenerativeModel('gemini-2.0-flash'); print('✅ 준비 완료')
```

## Step 1: 챗봇 설정 (✏️)

```python
CHATBOT_CONFIG={
    'subject':'✏️ [과목: 수학/영어/과학 등]',
    'grade':'✏️ [학년: 중1/고2 등]',
    'persona':'친절하고 명쾌한 과외 선생님',
}
print('✅ 챗봇 설정 완료')
```

## Step 2: 멀티턴 대화 실행

```python
chat=model.start_chat(history=[])
system_msg=f"""당신은 {CHATBOT_CONFIG['grade']} {CHATBOT_CONFIG['subject']} {CHATBOT_CONFIG['persona']}입니다.
학생의 질문에 쉽고 친절하게 답하고, 필요하면 예시를 들어 설명하세요."""
print('🎓 AI 학습 도우미 시작 (quit으로 종료)')
print('-'*40)
qa_log=[]
while True:
    q=input('학생: ')
    if q.lower() in ['quit','종료','exit']: break
    if not q.strip(): continue
    full_q=f"{system_msg}\n\n학생 질문: {q}" if not qa_log else q
    resp=chat.send_message(full_q)
    print(f'선생님: {resp.text}\n')
    qa_log.append({'질문':q,'답변':resp.text})
```

## Step 3: 대화 기록 저장

```python
import pandas as pd
if qa_log:
    pd.DataFrame(qa_log).to_csv('chat_log.csv',index=False,encoding='utf-8-sig')
    from google.colab import files; files.download('chat_log.csv')
    print(f'✅ 대화 {len(qa_log)}회 저장 완료')
else: print('대화 없음')
```
